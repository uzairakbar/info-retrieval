### Task 1
In this task, we will familiarize ourselves with some basic NumPy functionalities. Make sure that it is imported to your IPython shell.

```python
import numpy as np
```

**a)** Generate an arbitrary one dimensional array `n` with `n.shape=(8,)` and display it.

```python
n = np.array([3,8,12,18,7,11,30])
print(n)
```

**b)** Create a new array `n_odd` which consists of the odd entries of `n`. Display it.

```python
n_odd = n[1::2]
print(n_odd)
```

**c)** Create a new array `n_rev` which contains the entries of `n` in reverse order. Display it.

```python
n_rev = n[::-1]
print(n_rev)
```

**d)** What would be the output of the following code?

```python
a = np.array([1, 2, 3, 4, 5])
b = a[1:4]
b[0] = 200
print(a[1])
```

**e)** Create a two-dimensional array called `m` with `m.shape=(3,4)` and display it.

```python
m = np.array([[11, 12, 13, 14], [21, 22, 23, 24], [31, 32, 33, 34]])
print(m)
```

**f)** Create another array `m_revrowel` which contains the same elements as `m`, but with reversed rows. Display it.

```python
m_revrowel = m[:, ::-1]
print(m_revrowel)
```

**g)** Create yet another array `m_revall` which contains the same elements as `m`, but with reversed rows and columns. Display it.

```python
m_revall = m[::-1, ::-1]
print(m_revall)
```

**h)** Create an array `m_cut` which contains only the elements from the first and last rows and columns of `m`. Display it.

```python
m_cut = m[(0,-1),:][:,(0,-1)]
print(m_cut)
```
